# Creating new column for extracting compounds present in the formula it will helps us in acheving faster computation while searching for specifc compound, multiple compounds using text based indexing in mongodb

```python
def split_compounds(string):
    result = set()
    temp = ''
    for x in string:
        curr = ord(x)
        if 64 < curr < 91:
            result.add(temp)
            temp = ''
            temp += x
        elif 96 < curr < 123:
            temp += x   
    result.add(temp)
    result.remove('')
    return ' '.join(result), len(result)
        
```
